```python
df = pd.read_csv('csv/57_prepared2.csv')
```

```python
print(df.shape)
conv_id = [0,79]
df = df[df["Conversion page ID"].isin(conv_id)]
df = df.sort_values(["customer_id", "date"]).reset_index(drop=True)
df["path_no"] = df.groupby("customer_id")["conversion"].shift(1).fillna(0)#.cumsum()#.shift(1).fillna(0)
df["path_no"] = df.groupby("customer_id")["path_no"].cumsum()+ 1
df = df[df['path_no']==1]
df = df[['customer_id', 'date', 'Insertion name', 'Action type']]
print(df.shape)
```

```python
df_exp = df[df['Action type'].isin(['impression','event'])]
```

```python
df_need = df_exp.groupby(['customer_id','Action type']).count()
df_need = pd.DataFrame(df_need['date']).reset_index()
df_need
```

```python
df_need["path_no2"] = df_need["customer_id"].shift(1).fillna(0)
path = []
list1 = list(df_need["customer_id"])
list2 = list(df_need["path_no2"])
i=0
while i!=(df_need.shape[0]-1):
    if list1[i]==list1[i+1]:
        path.append(int(2))
        path.append(int(2))
        i+=2
    else:
        path.append(int(1))
        i+=1
else:
    path.append(int(1))
df_need["path_no2"] = path
df_need
```

```python
df_imp1 = df_need[(df_need['Action type']=='impression')]
df_ev1 = df_need[(df_need['Action type']=='event')]
```

```python
df_imp1.sort_values(['customer_id'], inplace=True)
df_ev1.sort_values(['customer_id'], inplace=True)
```

```python
df_imp1=df_imp1[['date']]
df_imp1['conv'] = 1
df_imp1 = df_imp1.groupby('date').sum().reset_index()
df_imp1['conv'] = df_imp1['conv']/df_imp1.conv.sum()
#
```

```python
df_ev1=df_ev1[['date']]
df_ev1['conv'] = 1
df_ev1 = df_ev1.groupby('date').sum().reset_index()
df_ev1['conv'] = df_ev1['conv']/df_ev1.conv.sum()
#
```

```python
df_imp1.conv = df_imp1['conv'].cumsum()
df_ev1['conv'] = df_ev1.conv.cumsum()
```

```python
df_imp1.to_csv('imp57.csv')
df_ev1.to_csv('ev57.csv')
```
